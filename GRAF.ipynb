```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
!pip install handcalcs
import handcalcs.render 

df=pd.read_excel('C:/Users/E1281290/Downloads/GRAFICOS_EXPERIMENTO_3/dados_exp_3.xlsx')

import numpy as np
import matplotlib.pyplot as plt

# Configuração para renderizar gráficos em SVG
%config InlineBackend.figure_format = 'svg'

# Constantes
V_0 = 4.7 # Tensão inicial em volts
R = 10000  # Resistência em ohms
C = 0.01 # Capacitância em farads

# Array de tempo de 0 a 5 segundos, com 50 pontos
t = np.linspace(0,2490, 100)

# Tensão no capacitor durante o carregamento
V_t = V_0 * (1 - np.exp(-t / (R*C)))

# Plotando o gráfico
plt.figure(figsize=(15, 7))
plt.plot(t, V_t,color='blue')
a='Plotagem dos valores de Tensão versus tempo'
#plt.title(f'{a.upper()}: valores experimentais e comparativo com o grafico ideal de um Capacitor')
plt.xlabel('Tempo (s)')
plt.ylabel('Tensão (V)')
plt.grid(False)

# Inserindo a equação V_0 dentro da figura
equation_text = r'$V_{(t)} = V_0 \left(1 - e^{-\frac{t}{RC}}\right)$'


plt.text(0.19, 0.1, equation_text, fontsize=16, transform=plt.gca().transAxes, color='blue');

x=df['Tempo']
y=df['Tensão']
plt.plot(x, y, 'o', color='red', markersize=7)

plt.annotate(
    'Pontos Coletados Experimentalmente\ntendendo a $5V$', 
    xy=(1500, 4.6), 
    xytext=(2000, 0.9), 
    arrowprops=dict(facecolor='red', arrowstyle='->', color='red'), 
    fontsize=13, 
    ha='center', color='red'
)


plt.annotate(
    'Carregamento Ideal do Capacitor \n gerado pela Equação $V_{(t)}$ \n Para efeitos de comparação: ', 
    xy=(100, 3), 
    xytext=(650, 1), 
    arrowprops=dict(facecolor='orange', arrowstyle='->', color='blue'), 
    fontsize=13, 
    ha='center', color='blue'
)


plt.savefig('grafico_alta_resolucap.svg', format='svg')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('text', usetex=False)


plt.figure(figsize=(12, 6))

t = np.linspace(0, 1940, 100)
dv_dt = 3.0 * np.exp(-0.01 * t)

plt.plot(t, dv_dt, color='blue')

plt.ylabel(r'$\frac{dv}{dt} \, \left(\frac{Volts}{min}\right)$', fontsize=15)
plt.xlabel('Tempo (s)', fontsize=15)

equation_text = r'$V_{(t)} = V_0 \left(1 - e^{-\frac{t}{RC}}\right)$'

# Anotações no gráfico
plt.annotate(
    r'Gráfico da Derivada da função:',
    xy=(1000, 2.9), 
    fontsize=13, 
    ha='center', 
    color='blue'
)
plt.annotate(
    equation_text,
    xy=(1000, 2.7), 
    fontsize=13, 
    ha='center', 
    color='blue'
)

plt.annotate(
    'Ponto no qual a variação de tensão em função do tempo é máxima, \n ou seja, se a tensão varia rapidamente com o tempo \n nos terminais do capacitor, ao multiplicar pela capacitância $C$, \n o valor numérico é igual ao da corrente em cada instante \n de modo que:', 
    xy=(0, 3), 
    xytext=(1160, 0.5), 
    arrowprops=dict(facecolor='red', arrowstyle='->', color='red'), 
    fontsize=13, 
    ha='center', 
    color='red'
)

plt.annotate(
    r'$ \lim_{t \to \infty} C \frac{d}{dt}(V_{t}) = 0 $',
    xy=(1600, 0.29), 
    fontsize=14, 
    ha='center', 
    color='black'
)

plt.savefig('ráfico da Derivada da função.svg', format='svg')
```
