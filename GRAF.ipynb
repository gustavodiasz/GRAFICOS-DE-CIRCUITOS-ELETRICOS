```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
!pip install handcalcs
import handcalcs.render 

df=pd.read_excel('C:/Users/E1281290/Downloads/GRAFICOS_EXPERIMENTO_3/dados_exp_3.xlsx')

import numpy as np
import matplotlib.pyplot as plt

# ConfiguraÃ§Ã£o para renderizar grÃ¡ficos em SVG
%config InlineBackend.figure_format = 'svg'

# Constantes
V_0 = 4.7 # TensÃ£o inicial em volts
R = 10000  # ResistÃªncia em ohms
C = 0.01 # CapacitÃ¢ncia em farads

# Array de tempo de 0 a 5 segundos, com 50 pontos
t = np.linspace(0,2490, 100)

# TensÃ£o no capacitor durante o carregamento
V_t = V_0 * (1 - np.exp(-t / (R*C)))

# Plotando o grÃ¡fico
plt.figure(figsize=(15, 7))
plt.plot(t, V_t,color='blue')
a='Plotagem dos valores de TensÃ£o versus tempo'
#plt.title(f'{a.upper()}: valores experimentais e comparativo com o grafico ideal de um Capacitor')
plt.xlabel('Tempo (s)')
plt.ylabel('TensÃ£o (V)')
plt.grid(False)

# Inserindo a equaÃ§Ã£o V_0 dentro da figura
equation_text = r'$V_{(t)} = V_0 \left(1 - e^{-\frac{t}{RC}}\right)$'


plt.text(0.19, 0.1, equation_text, fontsize=16, transform=plt.gca().transAxes, color='blue');

x=df['Tempo']
y=df['TensÃ£o']
plt.plot(x, y, 'o', color='red', markersize=7)

plt.annotate(
    'Pontos Coletados Experimentalmente\ntendendo a $5V$', 
    xy=(1500, 4.6), 
    xytext=(2000, 0.9), 
    arrowprops=dict(facecolor='red', arrowstyle='->', color='red'), 
    fontsize=13, 
    ha='center', color='red'
)


plt.annotate(
    'Carregamento Ideal do Capacitor \n gerado pela EquaÃ§Ã£o $V_{(t)}$ \n Para efeitos de comparaÃ§Ã£o: ', 
    xy=(100, 3), 
    xytext=(650, 1), 
    arrowprops=dict(facecolor='orange', arrowstyle='->', color='blue'), 
    fontsize=13, 
    ha='center', color='blue'
)


plt.savefig('grafico_alta_resolucap.svg', format='svg')
```
